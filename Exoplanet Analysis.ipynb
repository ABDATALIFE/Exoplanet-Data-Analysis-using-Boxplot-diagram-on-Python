```python
# Importing Libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Importing Dataset
dataset = pd.read_csv('Planets.csv')
dataset = dataset.sample(frac = 0.05)
```

```python
dataset.shape
```

```python
fig , ax = plt.subplots(figsize=(7,6))
ax.set_xscale("log") ## Changing the scale to logarithmic because the distance are too big
sns.boxplot(x = 'st_mass',y='discoverymethod',data =dataset)

sns.swarmplot(x = 'st_mass',y='discoverymethod',data =dataset,size=2,color='.3',linewidth = 0)

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True,left=True)
```
